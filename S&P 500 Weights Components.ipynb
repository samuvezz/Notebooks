This script retrievs data from the website https://www.slickcharts.com/sp500 through the HTML reading of the website. 

I was interested in the weights of the compamy weights composing the SP500 index on a daily basis

```python
import os.path
import csv

import requests
from bs4 import BeautifulSoup 
import pandas as pd
```

```python
url = 'https://www.slickcharts.com/sp500'
```

```python
header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}

#we use this code to pretend to be a browser. Otherwise this errror appears: HTTP Error 403: Forbidden

r = requests.get(url, headers=header)

df = pd.read_html(r.text)[0]
df = df.drop(df.columns[0], axis=1) 
```

```python
df['Date'] = pd.Timestamp.today().strftime('%d-%m-%Y')
#df.set_index('Date') in case you want the date to be the index, change the setting of df in the next line accordingly
df[['Date','Company','Symbol', 'Weight', 'Price', 'Chg', '% Chg']] #rearrange te columns
```

The following piece of code attaches the new retrived data to an Excel file so the final result would be a csv file with all the
SP500 data weights day by day

```python
#df = pd.DataFrame(df, columns=['Date','Company','Symbol', 'Weight', 'Price', 'Chg', '% Chg'])


print(df)
filename="S&P 500 Weights Components.csv"
header=True
if os.path.isfile(filename):
    header=False
df.to_csv(filename, mode="a", header=header, sep=",", quoting=csv.QUOTE_NONE, quotechar="",  escapechar="\\") 
```
